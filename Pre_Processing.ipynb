```python
df = dd.copy()
df.dropna(subset=['hashtags'], inplace=True)
df = df[['text']] 
df.columns = ['tweets']
df.head()
```

Pre-processing the tweets using NLTK

```python
# source  https://github.com/itsuncheng/text_preprocessing_with_NLTK/blob/master/preprocessing.ipynb
import nltk
from nltk.stem.wordnet import WordNetLemmatizer


nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
stop_words = nltk.corpus.stopwords.words(['english'])

print(stop_words)
```

```python
from nltk.tokenize import TweetTokenizer
from nltk.stem.wordnet import WordNetLemmatizer
lem = WordNetLemmatizer()

def cleaning(data):

    #removeing  hashtags
    tweet_without_hashtag = re.sub(r'#\w+', ' ', data)

    #removing urls
    tweet_without_url = re.sub(r'http\S+',' ', tweet_without_hashtag)

    #3. Removing mentions 
    tweet_without_mentions = re.sub(r'@\w+',' ', tweet_without_url)
    

    #Removing characters that are not in the English alphabets
    precleaned_tweet = re.sub('[^A-Za-z]+', ' ', tweet_without_mentions)

    #2. Tokenizing
    tweet_tokens = TweetTokenizer().tokenize(precleaned_tweet)

    #4. Removing Stopwords
    tokens_without_sw = [t for t in tweet_tokens if t not in stop_words]

    #3. Removing Punctuations
    tokens_without_punc = [w for w in tokens_without_sw if w.isalpha()]

    #5. lemmatisation
    text_cleaned = [lem.lemmatize(t) for t in tokens_without_punc]

    #6. Joining
    return " ".join(text_cleaned)
```
